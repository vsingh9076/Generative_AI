```python
def get_video_transcript(video_url):
    try:
        # Create a YouTube object
        yt = YouTube(video_url)

        # Get the automatic captions (if available)
        captions = yt.captions

        # Get the English captions if available, otherwise get the first available
        if 'en' in captions:
            caption = captions['en']
        else:
            caption = list(captions.values())[0]

        # Download the captions as a string
        transcript = caption.generate_srt_captions()

        return transcript

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
from google.cloud import speech_v1p1beta1 as speech
from google.cloud.speech_v1p1beta1 import types

def transcribe_video(video_path):
    client = speech.SpeechClient()

    with open(video_path, 'rb') as audio_file:
        content = audio_file.read()

    audio = types.RecognitionAudio(content=content)
    config = types.RecognitionConfig(
        encoding=types.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=16000,
        language_code='en-US',  # Update the language code if necessary
    )

    response = client.recognize(config=config, audio=audio)

    text = ""
    for result in response.results:
        text += result.alternatives[0].transcript + " "

    return text.strip()

if __name__ == '__main__':
    video_path = "/content/1 HOUR of Objects vs Hydraulic Press! (NEW).mp4"  # Replace with the path to your video file
    result = transcribe_video(video_path)

    print("Transcription:")
    print(result)

```

```python
import PyPDF2

# Open the PDF file
with open('/content/Rhonda_Byrne_The_Secret_2006.pdf', 'rb') as file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(file)

    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)

    # Loop through all pages and extract text
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        text = page.extract_text()
        print(f"Page {page_num + 1}:\n{text}\n")
```

### GET EMBEDDINGS OF TEXT USING OPENAI's text-embedding-ada-002 MODEL

```python
!pip install sentence_transformers~=2.2.2

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('intfloat/e5-small-v2')

def get_embedding(text):
   text = text.replace("\n", " ")
   return model.encode(text, normalize_embeddings=True)
```

### Insert data in Pinecone index

```python
!pip install pinecone-client -q

from google.colab import userdata
key = userdata.get('pinecone_api_key')

import pinecone
pinecone.init(api_key=key, environment='gcp-starter')
```

```python
pinecone.create_index(name='product-search', dimension=384)
```

```python
index = pinecone.Index('product-search')
```

```python
embedding_df = pd.read_csv("embedded.csv",index_col=0)
embedding_df.head()
#embedding_df['NameDescriptionVector'] = embedding_df.NameDescriptionVector.apply(eval).apply(np.array)
```

```python
docs = embedding_df.to_dict("records")
# docs[:3]
```

### Search data from index

```python
question_list = []
for i,row in embedding_df.iterrows():
  question_list.append(
      (
        str(row['ProductID']),
        model.encode(row['ProductName']).tolist(),
        {
            'ProductName': row['ProductName'],
            'Price (INR)':row['Price (INR)']
        }
      )
  )
  if len(question_list)==50 or len(question_list)==len(embedding_df):
    index.upsert(vectors=question_list)
    question_list = []
```

```python
query = "Shoes for men under 1500"
xq = model.encode([query]).tolist()
result = index.query(xq, top_k=6, includeMetadata=True)
result
```
